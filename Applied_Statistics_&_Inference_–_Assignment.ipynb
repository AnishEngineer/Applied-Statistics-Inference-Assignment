Question 1: What are Type I and Type II errors in hypothesis testing?

Answer:
In hypothesis testing, decisions are made based on data, but sometimes these decisions can be incorrect.

A Type I Error occurs when we reject a true null hypothesis. This means we conclude that there is an effect or difference when actually there is none. It is also called a false positive.

A Type II Error occurs when we fail to reject a false null hypothesis. This means we miss detecting a real effect. It is also called a false negative.

These errors directly impact decision-making. For example, in medical testing, a Type I error may lead to incorrect diagnosis, while a Type II error may result in missing a serious disease.

Question 2: What is the P-value and how is it interpreted?

Answer:
The P-value measures the strength of evidence against the null hypothesis.

If the P-value is less than 0.05, we reject the null hypothesis, which means there is significant evidence.
If the P-value is greater than 0.05, we do not reject the null hypothesis, meaning there is not enough evidence.

In simple terms, a smaller P-value indicates stronger evidence that the observed result is not due to chance.

Question 3: Difference between Z-test and T-test

Answer:

Uses normal distribution	Uses t-distribution

In practice, Z-tests are used when data size is large and parameters are known, while T-tests are preferred for smaller samples.

Question 4: What is a confidence interval and margin of error?

Answer:
A confidence interval is a range of values within which the true population parameter is expected to lie.

For example, if we say we are 95% confident that a value lies between 50 and 60, it means we believe the true value falls in that range.

The margin of error determines how wide this interval is. A larger margin of error results in a wider interval, while a smaller margin gives a more precise estimate.

Question 5: Purpose and assumptions of ANOVA

Answer:
ANOVA (Analysis of Variance) is used to compare the means of more than two groups to determine if there is a significant difference between them.

Its main purpose is to test whether all group means are equal or at least one is different.

The assumptions of ANOVA are:

The data is normally distributed

The samples are independent

The variances across groups are equal

For example, it can be used to compare the performance of students in three different classes.

Question 6: Python program for one-sample Z-test

Answer:



```python
import numpy as np
from statsmodels.stats.weightstats import ztest

data = [50, 52, 48, 51, 49, 47, 53, 50]
z_stat, p_value = ztest(data, value=50)

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

Interpretation:
If the P-value is less than 0.05, we reject the null hypothesis. Otherwise, there is no significant difference.

Question 8: Central Limit Theorem using Python

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.exponential(scale=2, size=1000)

sample_means = []
for i in range(100):
    sample = np.random.choice(data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=20)
plt.title("CLT Demonstration")
plt.show()
```

Even if the original data is not normally distributed, the sample means tend to follow a normal distribution

Question 9: Confidence interval function

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def confidence_interval(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)

    t = stats.t.ppf(0.975, df=n-1)
    margin = t * (std / np.sqrt(n))

    lower = mean - margin
    upper = mean + margin

    print("Confidence Interval:", (lower, upper))

    plt.hist(data)
    plt.title("Data Distribution")
    plt.show()

data = [10, 12, 14, 13, 15, 16, 14]
confidence_interval(data)
```

Question 10: Chi-square goodness-of-fit test

Answer:

```python
import scipy.stats as stats

observed = [50, 30, 20]
expected = [40, 40, 20]

chi_stat, p_value = stats.chisquare(observed, expected)

print("Chi-square:", chi_stat)
print("P-value:", p_value)
```

Explanation:
If the P-value is less than 0.05, the observed and expected distributions are significantly different.
If the P-value is greater than 0.05, there is no significant difference.
